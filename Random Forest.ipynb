```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

# Plant Village

## Load Dataset

```python
import numpy as np

dataset = np.load("/content/drive/MyDrive/Data Mining/Project/PlantVillage.npz")

X_train = dataset["X_train"]
X_test = dataset["X_test"]
y_train = dataset["y_train"]
y_test = dataset["y_test"]
```

## Model Training

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(max_depth=10, max_features='sqrt', min_samples_split=2, min_samples_leaf=1, class_weight='balanced', random_state=12)
model.fit(X_train, y_train)
```

## Model Evaluation

```python
# Predict on the training data
y_train_pred = model.predict(X_train)

# Calculate and print the train accuracy
train_accuracy = accuracy_score(y_train, y_train_pred) * 100
print(f"Train Accuracy: {train_accuracy}%")

y_pred = model.predict(X_test)

# Evaluate the model accuracy
print(f"Test Accuracy: {accuracy_score(y_test, y_pred)*100}%")
```

```python
classes = ["healthy", "early_blight", "late_blight", "bacterial_spot", "mosaic_virus"]
print(classification_report(y_test, y_pred, target_names=classes))
```

```python
cm = confusion_matrix(y_test, y_pred, labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)
fig, ax = plt.subplots(figsize=(8, 6))
disp.plot(ax=ax)
plt.tight_layout()
```
